```python
# import libraries
import piplite
await piplite.install('seaborn')

import pandas as pd
import numpy as np


import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# open the file
url = r"project_1.csv"
df = pd.read_csv(url) #to loud file
df
```

```python
# Add Columns
df['Team'] = ['Paksi FC', 'Diósgyőr', 'Paksi FC', 'Paksi FC', 'Ferencváros', 'MTK Budapest', 'Diósgyőr', 'Nyíregyháza', 'Zalaegerszeg', 'Puskás AFC', 'Nyíregyháza']
df['Country'] = ['Hungary', 'Spain', 'Hungary', 'Hungary', 'Hungary', 'Hungary', 'Hungary', 'Hungary', 'Hungary', 'Switzerland', 'Greece']
df['Position'] = ['CB', 'CDM', 'RB', 'CM', 'CDM', 'CB', 'RB', 'CB', 'RB', 'RB', 'CB']
df['Sub_position'] = ['RB', 'CB', 'RM-RW', 'CDM-CAM', 'CB-CM', 'CDM-RB', 'RM-RW', 'LB', 'RM-CAM', 'RM-LB', 'None']
df['Age'] = [32, 33, 29, 32, 27, 24, 30, 23, 25, 29, 27]
df['Hight (cm)'] = [187, 191, 177, 185, 190, 181, 181, 189, 183, 180, 185]
df['Market Value (k)'] = [450, 250, 650, 700, 700, 600, 600, 280, 400, 900, 380]

df
```

```python
# Statistics for Top 11 Players have best Definsive Stats
df.describe()
```

```python
# Market Values and Ages for Top 11 Players have best Definsive Stats

plt.scatter(df['Market Value (k)'], df['Age'])
for i in range(len(df)):
    label = f"{df['Name'].iloc[i]} ({df['Position'].iloc[i]})"
    plt.text(df['Market Value (k)'].iloc[i] + 10, df['Age'].iloc[i], label, fontsize=8)

plt.title('Value vs Age for Top 11 Players have best Definsive Stats')
plt.ylabel('Age')
plt.xlabel('Market Value (K)')

plt.show()
```

```python
# Passes Accuracy Distribution
sns.boxplot(df['Passes_accuracy'], fill=True)
plt.title('Distribution of Passes Accuarcy')

plt.show()
```

```python
# Hight Distribution
sns.boxplot(df['Hight (cm)'], fill=True, color='red')
plt.title('Distribution of Hight (cm)')

plt.show()
```

```python
# Passes Stats for Top 11 Players have best Definsive Stats

plt.figure(figsize=(15,5))
plt.scatter(df['Accurate_passes'], df['Passes_accuracy'])
for i in range(len(df)):
    label = f"{df['Name'].iloc[i]} ({df['Position'].iloc[i]})"
    plt.text(df['Accurate_passes'].iloc[i] + 4, df['Passes_accuracy'].iloc[i], label, fontsize=8)

plt.title('Passes Stats of Top 11 Players have best Definsive Stats')
plt.xlabel('Accurate Passes')
plt.ylabel('Passes Accuracy')

plt.show()
```

```python
# Select top 11 players by total defensive actions

df['Def_Actions'] = df['Tackles'] + df['Interceptions'] + df['Clearances']
top_def = df.sort_values('Def_Actions', ascending=False).head(11)

# Plot stacked bar
plt.bar(top_def['Name']+' ('+top_def['Position']+')', top_def['Tackles'], label='Tackles')
plt.bar(top_def['Name']+' ('+top_def['Position']+')', top_def['Interceptions'], 
        bottom=top_def['Tackles'], label='Interceptions')
plt.bar(top_def['Name']+' ('+top_def['Position']+')', top_def['Clearances'], 
        bottom=top_def['Tackles'] + top_def['Interceptions'], label='Clearances')

plt.xticks(rotation=45, ha='right')
plt.xlabel('Player')
plt.ylabel('Defensive Actions')
plt.title('Top 11 Players - Defensive Contributions')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Market Value by Position

plt.figure(figsize=(10,6))

avg_interceptions = df.groupby('Position')['Interceptions'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
plt.bar(avg_interceptions.index, avg_interceptions.values, color="green")
plt.xticks(rotation=45, ha="right")

plt.xlabel('Position')
plt.ylabel('Average Interceptions')
plt.title('Average Interceptions by Position')
plt.tight_layout()
plt.show()
```
